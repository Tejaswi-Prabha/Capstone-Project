```python
import requests
# Input from the user
symbol = input("Enter a stock ticker (e.g., AAPL): ").upper()
function = input("Enter a function (e.g., TIME_SERIES_INTRADAY, GLOBAL_QUOTE): ")

# Construct the URL using f-strings
url = f"https://www.alphavantage.co/query?function={function}&symbol={symbol}&apikey={API_KEY}"

# Make the API request
try:
    r = requests.get(url)
    r.raise_for_status()  # Raise an error for bad status codes (e.g., 404, 500)
    results = r.json()

    # Check if the response contains an error message
    if "Error Message" in results:
        print("Error:", results["Error Message"])
    else:
        print(results)  # Print the results

except requests.exceptions.RequestException as e:
    print("An error occurred:", e)
```

```python
# prompt: create a table

import pandas as pd

# Assuming 'report1', 'report2', etc., are already defined as in your original code

# Create a list to store the DataFrames
report_dfs = []

# Loop through the reports
for i in range(16):  # Adjust the range if you have more reports
  try:
    report_data = Annual_report[i] # Access the report data
    df = pd.DataFrame(report_data, index=[0]) # Create DataFrame for each report
    report_dfs.append(df)
  except (IndexError, KeyError) as e:
    print(f"Error processing report {i+1}: {e}")


# Concatenate all the DataFrames into a single DataFrame
if report_dfs:
  final_df = pd.concat(report_dfs, ignore_index=True)

  # Display or save the final DataFrame
  print(final_df)
  final_df.to_csv('combined_reports.csv', index=False)  # Save to CSV
else:
  print("No valid report data found.")
```
